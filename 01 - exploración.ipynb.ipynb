```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def jump():
  print("-" * 80, "\n" * 4)

def analizar_csv(nombre_archivo):
    try:
        # Leer el archivo CSV usando pandas
        df = pd.read_csv(nombre_archivo)

        # Mostrar el nombre de las columnas
        print(df.columns)
        jump()

        # Mostrar la información (# registros, # columnas, tipo dato, datos nulos)
        print(df.info())
        jump()

        # Mostrar los primeros 20 registros
        print(df.iloc[:20])
        jump()

        # Mostrar los ultimos 5 registros
        print(df.tail())
        jump()

        # Mostrar información estadística básica
        print("\nInformación estadística:")
        print(df.describe())
        jump()

        # Obtener las columnas numéricas
        columnas_numericas = df.select_dtypes(include=['number'])
        print(columnas_numericas)
        jump()


    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo}' no se encontró.")
    except pd.errors.ParserError:
        print(f"Error: No se pudo analizar el archivo '{nombre_archivo}'. Verifica que sea un archivo CSV válido.")
    except Exception as e:
        print(f"Error inesperado: {e}")

# Ejemplo de uso:
analizar_csv('/content/train.csv') # Reemplaza 'your_file.csv' con el nombre de tu archivo
```
